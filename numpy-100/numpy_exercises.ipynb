#### Write a function that takes a 1d numpy array and computes its reverse vector (last element becomes the first).

```python
def reverse(array):
    return np.flip(array)

a = np.random.uniform(0, 100, 10)
print(a)
print(reverse(a))
```

#### Given the following square array, compute the product of the elements on its diagonal. [[1 3 8] [-1 3 0] [-3 9 2]]


```python
a = np.array([[1, 3, 8], [-1, 3, 0], [-3, 9, 2]])
print(a.diagonal().prod())
```

#### Create a random vector of size (3, 6) and find its mean value.

```python
a = np.random.uniform(1, 100, (3, 6))
print(a)
print(a.mean())
```

#### Given two arrays a and b, compute how many time an item of a is higher than the corresponding element of b.
##### a: [[1 5 6 8] [2 -3 13 23] [0 -10 -9 7]]
##### b: [[-3 0 8 1] [-20 -9 -1 32] [7 7 7 7]]


```python
a = np.array([[1, 5, 6, 8], [2, -3, 13, 23], [0, -10, -9, 7]])
b = np.array([[-3, 0, 8, 1], [-20, -9, -1, 32], [7, 7, 7, 7]])
print(np.count_nonzero(a > b))
```

#### Create and normalize the following matrix (use min-max normalization). [[0.35 -0.27 0.56] [0.15 0.65 0.42] [0.73 -0.78 -0.08]]

```python
a = np.array([[0.35, -0.27, 0.56], [0.15, 0.65, 0.42], [0.73, -0.78, -0.08]])
norm = (a - a.min()) / (a.max() - a.min())
print(norm)
```
